## Sympy : 문자를  symbol화 해서 미적분 공식을 풀이하는데 유용

```python
import numpy as np
from sympy import symbols
from sympy import diff
from sympy import sin
from sympy import integrate
from sympy import limit
from sympy import Function, Eq, dsolve
```

```python
x=symbols('x') 
x*2
```

```python
expr=2*x
expr.subs(x,2)
```

```python
# differentiation
f=x**3
df1= diff(f,x) # f함수에서 x를 미분하는 의미
df2= diff(df1,x) # df1 에서 x를 미분
df3= diff(df2,x)
df3
```

```python
# function sin
f= sin(x)
df1=diff(f,x)
df1
```

```python
# integrate 적분
integrate(f,(x,3.14,2*3.14))
```

```python
# limit
limit(sin(x)/x,x,0)
```

```python
# 상미분 방정식을 코드로 풀어보기
t = symbols('t') # t 를 symbol화 시킴
x = Function('x') # x에 대한 Function 생성
ODE = Eq(x(t).diff(t,t)+x(t)) # 상미분 방정식 만듦
dsolve(ODE,x(t)) # 위의 ODE 방정식 풀이
```

## Scipy
http://docs.scipy.org/doc/scipy/reference/tutorial/index.html

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1.,0.8,0.4,0.3,0.2])

#plot 
#plt.plot(x,y,'*')

#interpolate
f_lin=interpolate.interp1d(x,y)
x_new=np.arange(1,5,0.1)
y_lin=f_lin(x_new)


# splint interpolation 곡선
tck=interpolate.splrep(x,y,s=0)
y_spl=interpolate.splev(x_new,tck,der=0)

#plot
fig, ax=plt.subplots()
ax.plot(x,y,'o',label='Data')
ax.plot(x_new,y_lin, label='Linear') # 직선
ax.plot(x_new,y_spl, label='Spline') # 곡선
ax.legend()
```
