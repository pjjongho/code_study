## Multiple Line Graphs - PyPlot Style

```python
x=np.linspace(0,2,100)
y1=0.5*x # linear
y2=0.5*x**2 # quadratic
y3=0.5*x**3 # cubic

plt.plot(x,y1,'b--')
plt.plot(x,y2)
plt.plot(x,y3)

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sibal')
```

## Decay sine example

```python
## decapy sine example
t=np.linspace(0,100,1000)
tau=100 # tau : 함수의 진폭을 표현하는 함수
y=np.sin(t)*np.exp(-t/tau)
plt.plot(t,y, 'b',label='Decay Oscillationg Response')
plt.ylabel('y[m]')
plt.xlabel('t[s]')
plt.legend()
```

```python
# %% -- Euler eq
t=np.linspace(0,1,100)
f=1 # frequency [Hz]
y_euler=np.exp(1j*2*np.pi*f*t)
y_cos=np.real(y_euler) # np.real : numpy 에서 복소수에서 실수를 추출하는 함수
y_sin=np.imag(y_euler) # np.imag : 복소수에서 허수 부분의 계수를 추출하는 함수

hfig, hax=plt.subplots()
hax.plot(t,y_cos,'-r',label='cos')
hax.plot(t,y_sin, '--b', label='sin')
hax.grid()
hax.legend()
```

```python
# 위의 cos, sin plot을 두 영역으로 나누어서 볼 수 있게 그리는 그래프

hfig, hax=plt.subplots(2,1) # plt.subplot(x,y) : subplot() -> () 안의 x,y 값을 입력하여 그래프를 몇 행열로 표시할건지 입력
hax[0].plot(t,y_cos,'-r',label='cos')
hax[1].plot(t,y_sin, '--b', label='sin')
hax[0].grid()
hax[1].grid()
hax[0].legend()
hax[1].legend()
```

## 3차원 그래프

```python
# %% 3D plot
# f(x,y)=sin(x)*sin(y)
import matplotlib.pyplot as plt


x=np.linspace(0,2*np.pi,20)
y=np.linspace(0,2*np.pi,20)
grid_x, grid_y=np.meshgrid(x,y) # meshgrid : 1차원 격자를 갖고 2차원 격자(grid)를 만들어줌

z=np.sin(grid_x)*np.sin(grid_y) 

hfig =plt.figure()
hax=hfig.add_subplot(projection='3d') # add_subplot : 3d 그래프로 하겠다는것을 반드시 명시해줘야함.
hax.plot_surface(grid_x,grid_y,z, cmap='jet') # cmap = 'jet' 다양한 색으로 표현
hax.set_xlabel('x')
hax.set_ylabel('y')
hax.set_zlabel('z')
```

```python
# %% Simple Animation

fig, ax = plt.subplots()

x=np.array([1,2,3,4,5])
y=np.array([1,1,1,1,1])

hp, = ax.plot(x,y)
ax.set_ybound([0,11])

for i in range(0,11,1): # 반복문 : 0부터 11까지 1씩 증가하는 정수들=i
    hp.set_ydata(i*y)
    plt.pause(0.5) # 0.5 초 마다 plot이 변하도록 설정
```
