```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
# Image size variables - 이미지 array를 3차원으로 생성
M=np.ones((2,3))
Height = 4
Width = 5
Depth = 3

# black
M=np.zeros((Height, Width, Depth))
plt.imshow(M) # plot 생성
plt.grid() # plt.grid()로 격자 생성
```

```python
# red
M[:,:,0]=255
plt.imshow(M)
plt.grid()
```

```python
# green
M=np.zeros((Height, Width, Depth))
M[:,:,1]=255
plt.imshow(M)
plt.grid()
```

```python
# blue
M=np.zeros((Height, Width, Depth))
M[:,:,2]=255
plt.imshow(M)
plt.grid()
```

```python
# White
M=np.ones((Height, Width, Depth))*255 # 모든원소들을 1*255로 하면 white 색상으로 나타남
plt.imshow(M)
plt.grid()
```

```python
# black
M=np.ones((Height, Width, Depth))*0 # 모든원소들을 0으로 만들면 black 색상으로 나타남
plt.imshow(M)
plt.grid()
```

```python
# change the top-left pixel to white # 첫번째 픽셀만 하얀색으로 바꾸고 싶을때
M[0,0,:]=255
plt.imshow(M)
plt.grid() # gird 안에 False를 입력하면 격자를 제거함
```

```python
# change the top-left pixel to green # 첫번째 픽셀만 녹색으로 바꾸고 싶을때
M[0,0,0]=0
M[0,0,1]=255
M[0,0,2]=0
plt.imshow(M)
```

```python
# change the top-left pixel to blue # 첫번째 픽셀만 파란색으로 바꾸고 싶을때
M[0,0,2]=255
M[0,0,1]=0
M[0,0,0]=0
plt.imshow(M)
plt.grid()
```

```python
# change the top-left pixel to red # 첫번째 픽셀만 적색으로 바꾸고 싶을때
M[0,0,0]=255
M[0,0,1]=0
M[0,0,2]=0
plt.imshow(M)
```

```python
# 이미지를 체크보드로 만들기
# Image & Matrix
# checkboard image
M=np.zeros((5,5,3))
plt.imshow(M)

# turn into white
M[:,:,:]=0
plt.imshow(M)

# checkboard
M[0::2,0::2,:]=1
plt.imshow(M)
M[1::2,1::2,:]=1
plt.imshow(M)
```

```python
# copy
N=M.copy()

# change white to gray
idx=np.where(N==1)
N[idx]=0.5
plt.imshow(N)
```

```python
# change black to white
idx=np.where(N<0.1)
N[idx]=1
plt.imshow(N)
```

```python
N=M.copy()
plt.imshow(N)

# color inverse
idx_w=np.where(N==1)
idx_b=np.where(N==0)
N[idx_w]=0
N[idx_b]=1
plt.imshow(N)
```

```python
# change white to green
idx1=np.where(N==1)
N[idx[0],idx1[1],0]=0
N[idx[0],idx1[1],2]=0
plt.imshow(N)
```

```python
# change black to red
flag_r=(N[:,:,0]==0)
flag_g=(N[:,:,1]==0)
flag_b=(N[:,:,2]==0)

flag_blk=flag_r&flag_g&flag_b
idx_blk=np.where(flag_blk==True)

N[idx_blk[0],idx_blk[1],0]=1
plt.imshow(N)
```

```python
# change black to red (another way)
N=M.copy()
plt.imshow(N)

N_sum=np.sum(N,axis=2)
idx_blk=np.where(N_sum==0)
N[idx_blk[0], idx_blk[1], 0]=1
plt.imshow(N)
```

```python
# change diagonal green
np.fill_diagonal(N[:,:,0],0)
np.fill_diagonal(N[:,:,1],1)
np.fill_diagonal(N[:,:,2],0)
plt.imshow(N)
```

```python
# jet scale
M=np.linspace(0,1,10)
M=M[:,np.newaxis]
M=np.repeat(M,10,axis=1)
plt.imshow(M,cmap='jet')
```

```python
# transpose
M_tr=np.transpose(M)
plt.imshow(M_tr, cmap='jet')
```

```python
dx=0.01
x1=np.arange(0,1.5,0.01)
y1=1/10*x1**2
```

```python
x2=np.arange(1.5,3+dx,dx)
N=50000
T=1
y2=np.zeros(x2.shape)
for n in range(N):
    if n%2==1:
        series=(4/(np.pi*n))*np.sin((2*np.pi*n*(x2-1.5))/T)
        y2=y2+series
```

```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Piecewise Function')
```
