## OpenCV (Computer Vision)

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import tkinter as tk
```

```python
# import image 이미지 파일 불러오기
im=cv.imread('son_pic.jpg')
im
```

```python
# BGR
plt.figure()
plt.imshow(im)
plt.title('Original')
```

```python
# RGB
rgb=cv.cvtColor(im, cv.COLOR_BGR2RGB)
plt.figure()
plt.imshow(rgb)
plt.title('RGB')
```

```python
# gray scale
gray=cv.cvtColor(im,cv.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(gray,cmap='gray')
plt.title('Gray')
```

```python
# blur
blur=cv.blur(gray,(10,10)) # 블러하는 박스 사이즈 설정
cv.cvtColor(blur,cv.COLOR_BGR2RGB)
plt.subplot(121) # '121' : 가로줄 한개(1), 새로줄 두개(2)를 넣고 싶은데 첫번째 이미지를 넣고 싶다는 의미
plt.imshow(gray, cmap='gray')
plt.title('GRAY')
plt.subplot(122)
plt.imshow(blur,cmap='gray')
plt.title('Blur')
```

```python
# edge detection
edges=cv.Canny(gray, 200,300) # 숫자는 영역을 조절하여 detect가 많이 되거나 적게 되게 설정할 수 있다.
plt.subplot(121) # '121' : 가로줄 한개(1), 새로줄 두개(2)를 넣고 싶은데 첫번째 이미지를 넣고 싶다는 의미
plt.imshow(gray, cmap='gray')
plt.title('Gray')
plt.subplot(122)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')
```
